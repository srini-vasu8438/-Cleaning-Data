Project - 1 (Mini Data Analyst Project: Survey & Insights)

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set style
sns.set(style="whitegrid")
```

```python
#Load CSV (upload file first in the left panel)
df = pd.read_csv('/content/Social Media Usage  (Responses) - Form responses 1.csv')  # update path if needed
df.head()
```

```python
# Check shape, info, and column names
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
df.info()
df.columns = [col.strip().replace('\n',' ').replace(' ','_') for col in df.columns]

# Display first few rows
df.head()
```

1️⃣ Age Group Distribution

```python
plt.figure(figsize=(6,4))
```

```python
sns.countplot(x='1._What_is_your_age_group?', data=df, palette='coolwarm')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Students')
plt.show()
```

2️⃣ Gender Distribution

```python
plt.figure(figsize=(5,4))
df['2._What_is_your_gender?'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue','pink'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

3️⃣ Support Reducing Social Media Usage

```python
plt.figure(figsize=(6,4))
sns.countplot(x='3._Would_you_support_reducing_your_social_media_usage_time_per_day?', data=df, palette='Set2')
plt.title('Support for Reducing Social Media Usage')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

4️⃣ Daily Usage Time

```python
plt.figure(figsize=(6,4))
sns.countplot(x='4._How_many_hours_do_you_spend_on_social_media_daily?', data=df, order=['Option 1Less than 1 hour', '1–3 hours','3–5 hours','More than 5 hours'])
plt.title('Daily Social Media Usage Time')
plt.xlabel('Hours per Day')
plt.ylabel('Number of Students')
plt.show()
```

5️⃣ Most Used Platform

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='5._Which_social_media_platform_do_you_use_most_often?', data=df, order = df['5._Which_social_media_platform_do_you_use_most_often?'].value_counts().index, palette='viridis')
plt.title('Most Used Social Media Platforms')
plt.xlabel('Number of Users')
plt.ylabel('Social Media Platform')
plt.show()
```

6️⃣ Main Reason for Using Social Media

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='6._What_is_your_main_reason_for_using_social_media?', data=df, order = df['6._What_is_your_main_reason_for_using_social_media?'].value_counts().index, palette='viridis')
plt.title('Main Reasons for Using Social Media')
plt.xlabel('Number of Users')
plt.ylabel('Reason')
plt.show()
```

7️⃣ Effect on Academic Performance

```python
plt.figure(figsize=(6,4))
sns.countplot(x='7._Do_you_think_social_media_affects_your_academic_performance?', data=df, palette='mako')
plt.title('Effect of Social Media on Academic Performance')
plt.xlabel('Impact')
plt.ylabel('Number of Students')
plt.show()
```

8️⃣ Follow Educational/Motivational Pages

```python
# Bar Chart for "Follow Educational/Motivational Pages"
educational_col_name = '8._Do_you_follow_educational_or_motivational_pages_on_social_media?'

plt.figure(figsize=(6,4))
df[educational_col_name].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Students Following Educational/Motivational Pages')
plt.xlabel('Response')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

9️⃣ Feeling After Long Usage

```python
# Bar Chart for "Feeling After Long Usage"
feelings_col_name = '9._How_do_you_feel_after_spending_a_long_time_on_social_media?'

plt.figure(figsize=(7,5))
df[feelings_col_name].value_counts().reindex(['Relaxed', 'Distracted', 'Stressed', 'Inspired']).plot(kind='bar', color=sns.color_palette('mako', 4))

plt.title('Feelings After Spending a Long Time on Social Media')
plt.xlabel('Feeling')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=20)
plt.show()
```

🔟 Primary Device Used

```python
# Bar Chart for "Primary Device Used"
device_col_name = '10._Which_device_do_you_primarily_use_for_accessing_social_media?'

plt.figure(figsize=(7,5))
df[device_col_name].value_counts().reindex(['Smartphone', 'Laptop', 'Tablet', 'Desktop Computer']).plot(kind='bar', color=sns.color_palette('viridis', 4))

plt.title('Primary Device Used for Accessing Social Media')
plt.xlabel('Device Type')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=20)
plt.show()
```

#SUMMERY FOR THIS PROJECT

### Project Summary – Social Media Usage Among Students

1. Majority of respondents (18–22 years) spend between 1–3 hours daily on social media.  
2. Instagram and YouTube are the most frequently used platforms among students.  
3. The primary reasons for using social media are *Entertainment* and *Communication*.  
4. Over 60% of students feel social media affects their academics — most report it negatively.  
5. Despite this, around 70% follow educational or motivational pages online.  
6. Most users feel *Distracted* or *Relaxed* after prolonged use, indicating mixed emotional effects.  
7. Smartphones are the dominant device for social media access, highlighting mobile dependency.  
8. Insight: Awareness programs on digital balance could help students use social media more productively.
