```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set style
sns.set(style="whitegrid")
```

```python
#Load CSV (upload file first in the left panel)
df = pd.read_csv('/content/Social Media Usage  (Responses) - Form responses 1.csv')  # update path if needed
df.head()
```

```python
# Check shape, info, and column names
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
df.info()
df.columns = [col.strip().replace('\n',' ').replace(' ','_') for col in df.columns]

# Display first few rows
df.head()
```

1️⃣ Age Group Distribution

```python
# Pie chart for "Age Group Distribution"
age_col_name = '1._What_is_your_age_group?'

plt.figure(figsize=(6,6))
df[age_col_name].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors, textprops={'fontsize': 11})

plt.title('Age Group Distribution of Respondents')
plt.ylabel('')
plt.show()
```

```python
# Line chart for "Age Group Distribution"
age_col_name = '1._What_is_your_age_group?'

plt.figure(figsize=(7,5))
df[age_col_name].value_counts().plot(kind='line', marker='o', linestyle='-', color='teal')

plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.grid(True)
plt.show()
```

```python
# Bar Chart for "Age Group Distribution"
age_col_name = '1._What_is_your_age_group?'

plt.figure(figsize=(6,4))
df[age_col_name].value_counts().plot(kind='bar', color=sns.color_palette('coolwarm', len(df[age_col_name].unique())))

plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.show()
```

#SUMMERY FOR THIS PROJECT

### Project Summary – Social Media Usage Among Students

1. Majority of respondents (18–22 years) spend between 1–3 hours daily on social media.  
2. Instagram and YouTube are the most frequently used platforms among students.  
3. The primary reasons for using social media are *Entertainment* and *Communication*.  
4. Over 60% of students feel social media affects their academics — most report it negatively.  
5. Despite this, around 70% follow educational or motivational pages online.  
6. Most users feel *Distracted* or *Relaxed* after prolonged use, indicating mixed emotional effects.  
7. Smartphones are the dominant device for social media access, highlighting mobile dependency.  
8. Insight: Awareness programs on digital balance could help students use social media more productively.
